## templat string

```python
from string import Template
```

```python
s=Template('selamat siang mr $s, apakah anda sudah $x ')
s.substitute(s='budi', x='siap')
```

```python
s=Template("seharusnya kita $action, sedari tadi sebelum $kejadian")
d={}
d['action']='balik badan'
d['kejadian']='ketahuan dia'
print(d)
```

```python
s.substitute(d)
```

```python
lebar=int(input("masukan lebar ="))
harga=10
item=lebar-harga

header_format='%-*s%*s'
format ='%-*s%*.2f'

print('=' * lebar)
print(header_format % (item, 'item', harga, 'harga'))
print('-' * lebar)

print(format % (item, 'apple', harga ,50))
print(format % (item, 'mangga', harga, 34))

print('=', lebar)
```

# String methods

### find metod

menemukan substring dalam string dengan mencari nilai indeks paling kiri dan mengambalikan nilai -1 jika tidak menemukan

```python
'selamat siang apa kabar, apakah anda siap hari ini'.find('anda')
```

```python
title="berkelana di atas bulan"
title.find('bulan')
```

```python
title.find('siapa')
```

```python
subjek="$$$ menjadi kaya sekarang $$$"
subjek.find('$$$')
```

```python
subjek.find('$$$',1)
```

```python
subjek.find('ng')
```
